```python
itens = []

def main():
  while True:
    print("\nMenu")
    print("1. Adicionar item")
    print("2. Listar itens")
    print("3. Marcar item como comprado")
    print("4. Remover item")
    print("5. Sair")

    escolha = input("Escolha uma opção: ")

    match escolha:
      case "1":
        item_nome = input("Nome do item:")
        item_quantidade = input("Quantidade:")
        item = {"nome": item_nome, "quantidade": item_quantidade, "comprado": False}
        adicionar_item(item)
      case "2":
        listar_itens()
      case "3":
        listar_itens()
        if itens:
          item = int(input("Digite o número do item que deseja marcar como comprado: "))
          marcar_item_comprado(item)
      case "4":
        listar_itens()
        if itens:
          item = int(input("Digite o número do item que deseja remover: "))
          remover_item(item)
      case "5":
        print("Saindo...")
        break
      case _:
        print("Opção inválida")

def adicionar_item(item):
  itens.append(item)
  print(f"Item '{item['nome']}' adicionado com sucesso!")

def listar_itens():
  print("Lista de Itens:")

  if not itens:
    print('Nenhum item encontrado')
    return

  for indice, item in enumerate(itens):
    status = "X" if item["comprado"] else " "
    print(f"{indice}. [{status}] {item['quantidade']} {item['nome']}")

def marcar_item_comprado(indice):
  if 0 <= indice < len(itens):
    itens[indice]["comprado"] = True
    print("Item comprado com sucesso!")
  else:
    print("Número inválido")

def remover_item(indice):
  if 0 <= indice < len(itens):
    itens.pop(indice)
    print("Item removido com sucesso!")
  else:
    print("Número inválido")
```

```python
main()
```
